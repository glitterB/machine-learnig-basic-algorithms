```python
#importing dependencies
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt("./Data/2_ex1.csv", delimiter=",", dtype=float)
data = np.insert(data,0,1,axis=1)
```

```python
x = data[:,0:2]
y = data[:,2].reshape(-1,1)
#plotting the data
fig,ax=plt.subplots()
ax.plot(x[:,1],y,"r.")
ax.set_xlabel("Population")
ax.set_ylabel("Cost")
ax.set_title("Linear Regression")
```

## Initialising Parameters

```python
theta = np.zeros([2,1])
iterations = 1500
alpha = 0.01
```

## Cost Function

$\sum_{i=0}^{m} \frac{1}{2m}\left(y_{actual}-y_{pred}\right)^2 $

```python
def compute_cost(x,y,theta):
    j = 0
    m = len(y)
    h = x@theta
    err = np.square(h-y)
    j = np.sum(err)/(2*m)
    return j
```

```python
print("Cost Function:",compute_cost(x,y,theta))
```

## Gradient Descent

### $\theta_i = \theta_i - \frac{\alpha}{m} \left(y_{actual}-y_{pred}\right)x_i$

```python
def gradient_descent(x,y,theta,alpha,iterations):
    m = len(y)
    j_hist = []
    
    for i in range(iterations):
        h = x@theta
        err = (h-y)
        j_vec = (x.T@err)*(alpha/m)
        theta -=j_vec
        
        j_hist.append(compute_cost(x,y,theta))
        
    return theta,j_hist
```

```python
theta,j_history = gradient_descent(x,y,theta,alpha,iterations)
```
