### Handwritten Digits Classification using Neural Network

- Feed Forward Algorithm with Pretrained _weights_ and _biases_

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing data
X = np.loadtxt("./Data/3_ex3data1.txt")
y = np.loadtxt("./Data/3_ex3data2.txt")

#importing pretrained weights
theta1 = np.loadtxt("./Data/3_theta1.txt")
theta2 = np.loadtxt("./Data/3_theta2.txt")
```

```python
#definig a function to display images
def img_from_vector(rv):
    """Given a row vector
    returns matrix of a image"""
    img = rv.reshape(20,20)
    return img.T
```

## Plotting sample _20x20_ Pixesl

```python
fig,ax = plt.subplots(10,10,figsize=(5,5))
for i in range(10):
    for j in range(10):
        ax[i][j].imshow(img_from_vector(X[np.random.randint(5000),:]),cmap="gray")
        ax[i][j].tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)
fig.suptitle("Handwritten Digits $20x20$")
```

Sigmoid Activation Function
### $g\left(z\right) = \frac{1}{1+e^{-z}}$

```python
#Activation Function
def sigmoid(z):
    g = np.zeros(np.shape(z));
    g = 1/(1+np.exp(-z));
    return g
```

### Neural Network
---
$a^{(1)} = X$

$z^{(2)} = \theta^Ta^{(1)}; a^{(2)} = g(z^{(2)})$

$z^{(3)} = \theta^Ta^{(2)}; a^{(3)} = g(z^{(3)})$

```python
def neural_network(X,w1,w2):
    #adding bias term to X (i.e. a1)
    a1 = np.insert(X,0,1,axis=1)
    
    z2 = np.dot(a1,w1.T)
    a2 = sigmoid(z2)
    
    #adding bias term to a2
    a2 = np.insert(a2,0,1,axis=1)
    z3 = np.dot(a2,w2.T)
    a3 = sigmoid(z3)
    
    return np.argmax(a3,axis=1)+1
```

```python
#Predictoin
fig2,ax2 = plt.subplots(5,10,figsize=(5,4))
for i in range(5):
    for j in range(10):
        inx = np.random.randint(5000)
        img = img_from_vector(X[inx,:])
        ax2[i][j].imshow(img,cmap='gray')
        p = neural_network(X[inx,:].reshape(1,400),theta1,theta2)
        ax2[i][j].set_title(p)
        ax2[i][j].tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)
```
