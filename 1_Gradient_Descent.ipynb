# Gradient Descent

- Gradient Descent is popular iterative optimization algorithm used to find Local Minimun.
- It is widely used in Machine Learning for optimizing Learnign Parameters.
- This Notebook will demonstrate how gradient descent works.

```python
#importing Libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#defining function

def f(x):
    return np.square(x+5)

#random data
x = np.linspace(-20,20,40)
y = f(x)

#plotting data
plt.figure(figsize=(5,3))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
```

### Deravative of Function can be calculated Numerically using Euler's Method

### $\frac{d}{dx}f(x) = \frac{f(x+h) - f(x)}{h}$

```python
#derivaitive of function
def d_x(x):
    """Euler's Method"""
    h = 0.01
    return (f(x+h)-f(x))/h
```
